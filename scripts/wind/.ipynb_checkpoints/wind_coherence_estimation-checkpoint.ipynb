```python
import matplotlib.pyplot as plt
import numpy as np
import scipy

from hdf5_sample_data import get_wind_samples_from_hdf5,\
    get_time_grid_from_hdf5, get_x_grid_from_hdf5
```

```python
HDF5_LO_RES = '../../data/wind/wind_samples_scalar_low-res_N5000.hdf5'
HDF5_HI_RES = '../../data/wind/wind_samples_scalar_high-res_N5000.hdf5'

# constants used for coherence evaluation:
#decay coefficients determine how fast coh. decays in each direction
C = [3.0, 3.0, 0.5] 
# constants determining mean wind velocity
U_STAR = 1.90       # shear velocity
Z_0 = 0.015         # roughness length
```

```python
#Load high- and low-res wind sample data
samples_lo = get_wind_samples_from_hdf5(HDF5_LO_RES, vector_field=False)
x_coords_lo = get_x_grid_from_hdf5(HDF5_LO_RES)
time_steps_lo = get_time_grid_from_hdf5(HDF5_LO_RES)
delta_t_lo = time_steps_lo[1] - time_steps_lo[0]

samples_hi = get_wind_samples_from_hdf5(HDF5_HI_RES, vector_field=False)
x_coords_hi = get_x_grid_from_hdf5(HDF5_HI_RES)
time_steps_hi = get_time_grid_from_hdf5(HDF5_HI_RES)
# delta_t_hi = time_steps_lo[1] - time_steps_lo[0] ## original
delta_t_hi = time_steps_hi[1] - time_steps_hi[0] ## modified
```

```python
def get_true_coherence(freq, x_coord_j, x_coord_k):
    '''
    Evaluate prescribed coherence function for given frequency and spatial coordinates
    '''
    mean_vel_j = 2.5 * U_STAR *np.log(x_coord_j[2] / Z_0)
    mean_vel_k = 2.5 * U_STAR *np.log(x_coord_k[2] / Z_0)
    numerator = freq * np.linalg.norm(np.multiply(C, x_coord_j - x_coord_k))
    denominator = np.abs(mean_vel_j) + np.abs(mean_vel_k)
    coherence = np.exp(-numerator / denominator)
    return coherence
```

```python
def get_empirical_coherence_from_csd(wind_samples_j, wind_samples_k, delta_t, 
                                     nperseg, N=None):
    '''
    Estimate coherence from wind samples at two different spatial coordinates. Uses formula C_jk = abs(P_jk)**2/(P_jj*P_kk) where P_jj and P_kk are power spectral density estimates for winds at coordinates j and k; and P_jk is cross spectral density estimate. P_** are calculated by averaging individual estimates over N samples before evaluating the coherence formula, this seems to be more stable.
    scipy splits signal into segments to estimate power spectral density, "nperseg" controls the length of each segment.
    '''
    # get frequency grid where coherence will be evaluated
    f_empirical, _ = scipy.signal.csd(wind_samples_j[0,  :], 
                                      wind_samples_j[0,  :], fs=1/delta_t, nperseg=nperseg)
    # Compute average CPSD over multiple samples
    N = len(wind_samples_j) if N is None else N
    Pjj_avg = np.mean([
        scipy.signal.csd(wind_samples_j[i, :], wind_samples_j[i, :], fs=1/
                         delta_t, nperseg=nperseg)[1]
        for i in range(N)
    ], axis=0)

    print(f'Pjj_avg shape: {Pjj_avg.shape}')
    Pkk_avg = np.mean([
        scipy.signal.csd(wind_samples_k[i, :], wind_samples_k[i, :], fs=1/
                         delta_t, nperseg=nperseg)[1]
        for i in range(N)
    ], axis=0)
    print(f'Pkk_avg shape: {Pkk_avg.shape}')
    Pjk_avg = np.mean([
        scipy.signal.csd(wind_samples_j[i, :], wind_samples_k[i, :], fs=1/
                         delta_t, nperseg=nperseg)[1]
        for i in range(N)
    ], axis=0)

    print(f'Pjk_avg shape: {Pjk_avg.shape}')

    coh = np.abs(Pjk_avg)**2 / (Pjj_avg * Pkk_avg)
    return f_empirical, coh
```

```python
def get_empirical_coherence_direct(wind_samples_j, wind_samples_k, delta_t, 
                                    nperseg, N=None):
    '''
    Estimate coherence from wind samples at two different spatial coordinates using scipy's coherence function. scipy splits signal into segments to estimate coherence, "nperseg" controls the length of each segment.
    '''
    f_empirical, _ = scipy.signal.coherence(wind_samples_j[0, 0, :], 
                                            wind_samples_j[0, 1, :],
                                                         fs=1/delta_t, nperseg=nperseg)

    # Compute average coherence over multiple samples
    N = len(wind_samples_j) if N is None else N
    coh_avg = np.mean([
            scipy.signal.coherence(wind_samples_j[i, 0, :], 
                                   wind_samples_k[i, 1, :], fs=1/delta_t, nperseg=nperseg)[1]
            for i in range(N)
        ], axis=0)
    return f_empirical, coh_avg
```

```python
# pick out two coords to evaluate coherence
# ind1 = 10 #2
# ind2 = 11 #6

ind1 = 2
ind2 = 6
x_1_lo = x_coords_lo[ind1, :]
vel_1_lo = samples_lo[:, ind1, :]
x_2_lo = x_coords_lo[ind2, :]
vel_2_lo = samples_lo[:, ind2, :]

x_1_hi = x_coords_hi[ind1, :]
vel_1_hi = samples_hi[:, ind1, :]
x_2_hi = x_coords_hi[ind2, :]
vel_2_hi = samples_hi[:, ind2, :]

print("x1lo", x_1_lo)
print("x2lo", x_2_lo)
print("x1hi", x_1_hi)
print("x2hi", x_2_hi)
```

```python
def compare_coherence(x1, x2, vel1, vel2, delta_t, npserseg, N):
    freq_lo, emp_coh_lo = get_empirical_coherence_from_csd(vel1, vel2,
                                                           delta_t, npserseg, N)
    #NOTE - empirical coherence is squared, so true coherence gets squared here
    true_coh_lo = [get_true_coherence(freq, x1, x2)**2 for freq in freq_lo]
    plt.figure()
    plt.plot(freq_lo, true_coh_lo, '-o', label='True - Low Res')
    plt.plot(freq_lo, emp_coh_lo, '-^', label='Empirical - Low Res')
    plt.xlabel('Frequency')
    plt.ylabel('Coherence')
    plt.legend()
    plt.show()
```

```python
#estimate low-res empirical coherence:
nperseg_lo = len(time_steps_lo)/2
N = 5000
compare_coherence(x_1_lo, x_2_lo, vel_1_lo, vel_2_lo, delta_t_lo, nperseg_lo, 
                  N)
```

```python
#estimate high-res empirical coherence:
nperseg_hi = len(time_steps_hi) / 2
N = 5000
compare_coherence(x_1_hi, x_2_hi, vel_1_hi, vel_2_hi, delta_t_hi, nperseg_hi, 
                  N)
```
