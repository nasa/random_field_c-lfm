# Constructing a dense evaluation grid

```python
def dense_eval(grid: RectangularGrid, discretization: tuple):
    '''
    Returns a (N**ndim, ndim) discretized grid over the domain. 
    '''
    assert len(discretization) == grid.ndim

    # Create a tuple of ranges for each dimension
    ranges = [torch.linspace(0.0, 1.0, d, device=grid.low.device) 
              for d in discretization]
    
    # Generate the coordinate matrices
    coords = torch.meshgrid(*ranges)
    
    # Stack the coordinates and reshape to (N**ndim, ndim)
    x = torch.stack(coords, dim=-1).reshape(-1, grid.ndim)

    # Scale and shift to the grid's domain
    low = grid.low.unsqueeze(0)
    high = grid.high.unsqueeze(0)
    x = x * (high - low) + low
    
    return x

'''
Grid ordering:
(x, y, z, t)
'''

disc = (1, 10, 10, 256) ## exact grid size of the data.
# disc = (1, 50, 150, 50) ## nice grid size for plotting
x = dense_eval(vae.grid, disc)
x = x[None, :, :].repeat(batch_size, 1, 1)

y = torch.linspace(vae.grid.low[1], vae.grid.high[1], disc[1])
z = torch.linspace(vae.grid.low[2], vae.grid.high[2], disc[2])
t_grid = torch.linspace(vae.grid.low[3], vae.grid.high[3], disc[3])
```

```python
with torch.no_grad():
    mu, logvar = vae.encode(u)
    # z = reparameterize(mu, logvar)
    z = torch.randn_like(mu)
    u_hat = vae.decode(z, x).cpu()
    
u_hat = u_hat.reshape(batch_size, *disc)
u_hat.shape, u.shape
```

```python
generated_energy = torch.square(u_hat).sum((1, 2, 3))
true_energy = torch.square(u).sum(1).cpu()

fig, ax = plt.subplots(1, 2, figsize = (10, 5))

ax[0].set_title('Energy of True Samples')
ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('Energy')

ax[1].set_title('Energy of Generated Samples')
ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('Energy')

for i in range(generated_energy.shape[0]):
    ax[0].plot(t_grid, true_energy[i,:], color = 'red', alpha = 0.05)
    ax[1].plot(t_grid, generated_energy[i,:], color = 'red', alpha = 0.05)
plt.show()
```

```python
## reshaping the dense x grid to be (batch x space x time x dim)
x_bar = x.reshape(batch_size, disc[1], disc[2], disc[3], vae.grid.ndim)
x_bar = x.reshape(batch_size, disc[1] * disc[2], disc[3], vae.grid.ndim)
## stripping off time coord
x_bar = x_bar[:, :, :, :3].cpu()
print('Coordinate shape:')
x_bar.shape
```

```python
print('Function output shape:')
u_bar = u_hat.reshape(batch_size, disc[1] * disc[2], disc[3]).cpu()
u_bar.shape
```

```python
fig, ax = plt.subplots(5, 5, figsize = (15, 10))

for i in range(ax.shape[0]):
    for j in range(ax.shape[1]):
        idx_1 = np.random.randint(x_bar.shape[1])
        idx_2 = np.random.randint(x_bar.shape[1])

        x1 = x_bar[0, idx_1, 0, :]
        x2 = x_bar[0, idx_2, 0, :]

        v1 = u_bar[:, idx_1, :]
        v2 = u_bar[:, idx_2, :]

        nperseg = v1.shape[-1] // 2
        coh = coherence(v1, v2, nperseg = nperseg)
        freq_grid = torch.fft.rfftfreq(nperseg, d = data.dt)
        true_coh = torch.tensor([get_true_coherence(f, x1, x2) ** 2 for f in freq_grid])

        ax[i, j].set_xlim(0.0, 3.0)
        ax[i, j].set_title(f'idx_1: {idx_1}, idx_2: {idx_2}')
        ax[i, j].plot(freq_grid, true_coh)
        ax[i, j].plot(freq_grid, coh)

fig.tight_layout()
plt.show()
```
