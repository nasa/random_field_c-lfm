```python
from pathlib import Path
import os
import torch
from torch.utils.data import DataLoader
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import simpson
from scipy.stats import gaussian_kde

from fgm.vae import FunctionalVAE
from fgm.problems.materials import MaterialsTrain, MaterialsVal
```

```python
device = torch.device(0)

version = 0
epoch = 24499
step = 171500

# variant = 'tb_residual'
variant = 'corrected_origin'
# variant = 'random_boundary'
name = 'epochs=50000-bs=128-num_colloc=500-lr=0.001-h_size=128-res_weight=1e-06-kld_weight=5e-06'
folder = Path(f'/Users/tashah/Desktop/functional-generative-modeling/results/materials/{variant}/{name}/version_{version}/')
model_path = folder / f'checkpoints/epoch={epoch}-step={step}.ckpt'

vae = FunctionalVAE.load_from_checkpoint(model_path).to(device)

val_metrics = ['val_corr_loss', 'val_loss_observable', 'val_loss_unobservable']
metrics = pd.read_csv(folder / 'metrics.csv').drop(val_metrics, axis = 1).dropna()

val_corr_loss =  pd.read_csv(folder / 'metrics.csv')['val_corr_loss'].dropna()
val_loss_observable = pd.read_csv(folder / 'metrics.csv')['val_loss_observable'].dropna()
val_loss_unobservable =  pd.read_csv(folder / 'metrics.csv')['val_loss_unobservable'].dropna()
```

```python
resolution = 25

train_data = MaterialsTrain()
test_data = MaterialsVal(x_sensor = train_data.X_u)
# test_loader = DataLoader(test_data, batch_size = 10000, shuffle = True)
test_loader = DataLoader(test_data, batch_size = 1000, shuffle = True)


u, x, f = next(iter(test_loader))
u, x = u.to(device), x.to(device)

X = test_data.X.unsqueeze(0).repeat(x.shape[0], 1, 1).to(device)


'''
Previously I was finding that samples generated using torch._randn_like(mu) produced very bad results.
However, since I corrected ** the residual calculation those samples arent looking as bad.
** The correction I made was to repeat the origin boundary point along the batch dimention.

This weakens the argument for using LFM slightly. 
However, LFM still provides the benifit of being very easy to incorporate conditioning information.
Additionally I do believe that as problems become increasingly complex we will need LFM to sample from p(z).
Otherwise most SOTA generative models would just use a VAE because that is much simpler.
'''
with torch.no_grad():
    mu, logvar = vae.encode(u)
    z = vae.reparameterize(mu, logvar)
    # z = torch.randn_like(mu)
    ux_hat, uy_hat, e_hat = vae.decode(z, X).chunk(3, dim = 2)
    
ux_hat = ux_hat.squeeze().reshape(-1, resolution, resolution).detach().cpu()
uy_hat = uy_hat.squeeze().reshape(-1, resolution, resolution).detach().cpu()
e_hat = e_hat.squeeze().reshape(-1, resolution, resolution).detach().cpu()

_, _, e_true = f.chunk(3, dim = 2)
e_true = e_true.reshape(-1, resolution, resolution).squeeze()

print(f'True E mean: {e_true.mean()}, Generated E mean: {e_hat.mean()}')
print(f'True E std: {e_true.std()}, Generated E std: {e_hat.std()}')

points_x = torch.linspace(0.0, 1.0, 25)

squared_diffs = (e_true.mean(dim = 0) - e_hat.mean(dim = 0)) ** 2
squared_true = e_true.mean(dim = 0) ** 2
true_L2 = simpson(simpson(squared_true, x = points_x), x = points_x)
diff_L2 = simpson(simpson(squared_diffs, x = points_x), x = points_x)
print(f'Mean Field Error: {diff_L2 / true_L2}')

squared_diffs = (e_true.std(dim = 0) - e_hat.std(dim = 0)) ** 2
squared_true = e_true.std(dim = 0) ** 2
true_L2 = simpson(simpson(squared_true, x = points_x), x = points_x)
diff_L2 = simpson(simpson(squared_diffs, x = points_x), x = points_x)
print(f'Std Field Error: {diff_L2 / true_L2}')
```

```python
## plotting a few samples for sanity check
num_samples = 3
fig, ax = plt.subplots(num_samples, 3, figsize = (15, num_samples * 5))

ax[0, 0].set_title('Generated ux')
ax[0, 1].set_title('Generated uy')
ax[0, 2].set_title('Generated e')
for i in range(num_samples):
    c = ax[i, 0].pcolor(ux_hat[i, :, :])
    fig.colorbar(c)
    c = ax[i, 1].pcolor(uy_hat[i, :, :])
    fig.colorbar(c)
    c = ax[i, 2].pcolor(e_hat[i, :, :])
    fig.colorbar(c)
fig.tight_layout()
plt.show()
```

```python
corr = True

y_line_A = 0.75
y_line_B = 0.50
y_line_C = 0.25

test_E = test_data.E
test_E = test_E[~(test_E > e_hat.max()).any(dim = 1), :]
# test_E = test_E[~(test_E > 1.98).any(dim = 1), :]
# test_E = test_E[~(test_E < 0.95).any(dim = 1), :]


if corr:
    # obtaining the emperical correlation function from the test data
    E_corr_A = torch.corrcoef(test_E[:, (test_data.X[:, 1] == y_line_A)].T)
    E_corr_B = torch.corrcoef(test_E[:, (test_data.X[:, 1] == y_line_B)].T)
    E_corr_C = torch.corrcoef(test_E[:, (test_data.X[:, 1] == y_line_C)].T)
else:
    E_corr_A = torch.cov(test_E[:, (test_data.X[:, 1] == y_line_A)].T)
    E_corr_B = torch.cov(test_E[:, (test_data.X[:, 1] == y_line_B)].T)
    E_corr_C = torch.cov(test_E[:, (test_data.X[:, 1] == y_line_C)].T)

## constructing a band of points at y = 0.75, 0.50, 0.25
A_points = torch.stack([
    torch.linspace(0.0, 1.0, 25),
    torch.full((25,), y_line_A),
], dim = 1).unsqueeze(0).repeat(z.shape[0], 1, 1).to(device)

B_points = torch.stack([
    torch.linspace(0.0, 1.0, 25),
    torch.full((25,), y_line_B)
], dim = 1).unsqueeze(0).repeat(z.shape[0], 1, 1).to(device)

C_points = torch.stack([
    torch.linspace(0.0, 1.0, 25),
    torch.full((25,), y_line_C)
], dim = 1).unsqueeze(0).repeat(z.shape[0], 1, 1).to(device)

## evaluating model at band of points
with torch.no_grad():
    _, _, e_A = vae.decode(z, A_points).cpu().chunk(3, dim = 2)
    _, _, e_B = vae.decode(z, B_points).cpu().chunk(3, dim = 2)
    _, _, e_C = vae.decode(z, C_points).cpu().chunk(3, dim = 2)

e_A = e_A.squeeze()
e_B = e_B.squeeze()
e_C = e_C.squeeze()

## compute the emperical correlation of the generated points on the band
points_x = A_points[0, :, 0].cpu()

if corr:
    e_corr_A = torch.corrcoef(e_A.T)
    e_corr_B = torch.corrcoef(e_B.T)
    e_corr_C = torch.corrcoef(e_C.T)
else:
    e_corr_A = torch.cov(e_A.T)
    e_corr_B = torch.cov(e_B.T)
    e_corr_C = torch.cov(e_C.T)

## finding the midpoint on the x axis of the data resolution
idx = int(25 / 2)
fig, ax = plt.subplots(1, 1)

ax.plot(points_x, E_corr_A[idx, :], label = 'True', color = 'black')
ax.plot(points_x, e_corr_A[idx, :], label = 'Generated (A)', color = 'red', linestyle = 'dashed')
ax.plot(points_x, e_corr_B[idx, :], label = 'Generated (B)', color = 'purple', linestyle = 'dashed')
ax.plot(points_x, e_corr_C[idx, :], label = 'Generated (C)', color = 'green', linestyle = 'dashed')
ax.legend()

# ax.set_ylim(0.65, 1.01)
ax.set_xlabel('x1')
ax.set_ylabel('C(x1, 0.5)')

if corr:
    ax.set_title('Correlation')
    fig.savefig('corr.png')
else:
    ax.set_ylim(0.01, 0.03)
    ax.set_title('Covariance')
    fig.savefig('cov.png')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 5))

c = ax[0].pcolor(points_x, points_x, (e_true.mean(dim = 0) - e_hat.mean(dim = 0)) )
fig.colorbar(c)
c = ax[1].pcolor(points_x, points_x, (e_true.std(dim = 0) - e_hat.std(dim = 0)) )
fig.colorbar(c)

ax[0].set_title('Error of Mean Fields')
ax[0].set_xlabel('x1')
ax[0].set_ylabel('x2')

ax[1].set_title('Error of Std Fields')
ax[1].set_xlabel('x1')
ax[1].set_ylabel('x2')
fig.tight_layout()
plt.show()
```
