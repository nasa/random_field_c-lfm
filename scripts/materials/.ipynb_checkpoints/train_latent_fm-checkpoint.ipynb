```python
from pathlib import Path
import os
import torch
from torch import nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
import pandas as pd
import numpy as np
from torchdyn.core import NeuralODE
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from tqdm import tqdm

from fgm.problems.materials import MaterialsTrain, MaterialsVal
from fgm.vae import FunctionalVAE
```

```python
device = torch.device(0)

epochs = 50000
bs = 128
lr = 0.001
h_size = 128
res_weight = 1e-6
kld_weight = 5e-6
emb_dim = 0
theta = 0.0

version = 0
epoch = 10749
step = 75250

name = f'epochs={epochs}-bs={bs}-lr={lr}-h_size={h_size}-res_weight={res_weight}-kld_weight={kld_weight}'
folder = Path(f'/Users/tashah/Desktop/functional-generative-modeling/results/materials/vae/{name}/version_{version}/')
model_path = folder / f'checkpoints/epoch={epoch}-step={step}.ckpt'

vae = FunctionalVAE.load_from_checkpoint(model_path).to(device)
metrics = pd.read_csv(folder / 'metrics.csv').drop('val_corr_loss', axis = 1).dropna()
val_corr_loss =  pd.read_csv(folder / 'metrics.csv')['val_corr_loss'].dropna()
```

```python
train_data = MaterialsTrain()
test_data = MaterialsVal(x_sensor = train_data.X_u)
train_loader = DataLoader(train_data, batch_size = 512, shuffle = True)

sigma = 0.01

h_size = 256
flow = nn.Sequential(
    nn.Linear(128 + 1, h_size),
    nn.GELU(),
    nn.Linear(h_size, h_size),
    nn.GELU(),
    nn.Linear(h_size, h_size),
    nn.GELU(),
    nn.Linear(h_size, h_size),

    nn.GELU(),
    nn.Linear(h_size, h_size),
    
    nn.GELU(),
    nn.Linear(h_size, 128),
).to(device)

opt = torch.optim.Adam(flow.parameters())

u, x = next(iter(train_loader))

hist = []

for epoch in tqdm(range(4000)):
    for batch in train_loader:

        opt.zero_grad()
        
        u, x = batch
        u, x = u.to(device), x.to(device)
        u_flat = u.flatten(start_dim = 1)
        
        with torch.no_grad():
            z1 = vae.reparameterize(*vae.encode(u_flat))
        
        z0 = torch.randn_like(z1)
        
        t = torch.rand(z1.shape[0], device = device).unsqueeze(-1)
        mu_t = t * z1 + (1.0 - t) * z0
        zt = mu_t + sigma * torch.randn_like(mu_t)
        vt = flow(torch.cat([zt, t], dim = -1))
        
        loss = F.mse_loss(vt, z1 - z0)
        loss.backward()
        opt.step()

        hist.append(loss.item())
        
plt.plot(hist) 
```

```python
class torch_wrapper(torch.nn.Module):
    '''
    https://github.com/atong01/conditional-flow-matching/tree/main
    Wraps model to torchdyn compatible format.
    '''
    def __init__(self, model):
        super().__init__()
        self.model = model
    
    def forward(self, t, x, *args, **kwargs):
        t = t.repeat(x.shape[0]).unsqueeze(-1)
        return self.model(torch.cat([x, t], dim = -1))

node = NeuralODE(
    torch_wrapper(flow),
    solver = 'dopri5',
    sensitivity = 'adjoint',
    atol = 1e-4, 
    rtol = 1e-4
).to(device)

num_samples = 5000
with torch.no_grad():
    traj = node.trajectory(
        torch.randn(num_samples, 128, device = device),
        t_span = torch.linspace(0.0, 1.0, 100, device = device)
    )
    
z = traj[-1, :, :]
```

```python
## decoding samples from flow matching

resolution = 25
X = test_data.X.unsqueeze(0).repeat(z.shape[0], 1, 1).to(device)

with torch.no_grad():
    ux, uy, e = vae.decode(z, X).chunk(3, dim = 2)
    
ux = ux.squeeze().reshape(-1, resolution, resolution).detach().cpu()
uy = uy.squeeze().reshape(-1, resolution, resolution).detach().cpu()
e = e.squeeze().reshape(-1, resolution, resolution).detach().cpu()

print(f'True E mean: {test_data.E.mean()}, Generated E mean: {e.mean()}')
print(f'True E std: {test_data.E.std()}, Generated E std: {e.std()}')
```

```python
N = 2
fig, ax = plt.subplots(N, 3, figsize = (15, N * 5))
ax[0, 0].set_title('ux')
ax[0, 1].set_title('uy')
ax[0, 2].set_title('e')
for i in range(N):
    c = ax[i, 0].pcolor(ux[i, :, :])
    fig.colorbar(c)
    c = ax[i, 1].pcolor(uy[i, :, :])
    fig.colorbar(c)
    c = ax[i, 2].pcolor(e[i, :, :])
    fig.colorbar(c)
fig.tight_layout()
plt.show()
```

```python
y_line_A = 0.75
y_line_B = 0.50
y_line_C = 0.25

## obtaining the emperical correlation function from the test data
E_corr_A = torch.corrcoef(test_data.E[:, (test_data.X[:, 1] == y_line_A)].T)
E_corr_B = torch.corrcoef(test_data.E[:, (test_data.X[:, 1] == y_line_B)].T)
E_corr_C = torch.corrcoef(test_data.E[:, (test_data.X[:, 1] == y_line_C)].T)

## constructing a band of points at y = 0.75, 0.50, 0.25
A_points = torch.stack([
    torch.linspace(0.0, 1.0, 25),
    torch.full((25,), y_line_A),
], dim = 1).unsqueeze(0).repeat(z.shape[0], 1, 1).to(device)

B_points = torch.stack([
    torch.linspace(0.0, 1.0, 25),
    torch.full((25,), y_line_B)
], dim = 1).unsqueeze(0).repeat(z.shape[0], 1, 1).to(device)

C_points = torch.stack([
    torch.linspace(0.0, 1.0, 25),
    torch.full((25,), y_line_C)
], dim = 1).unsqueeze(0).repeat(z.shape[0], 1, 1).to(device)

## evaluating model at band of points
with torch.no_grad():
    _, _, e_A = vae.decode(z, A_points).cpu().chunk(3, dim = 2)
    _, _, e_B = vae.decode(z, B_points).cpu().chunk(3, dim = 2)
    _, _, e_C = vae.decode(z, C_points).cpu().chunk(3, dim = 2)

## compute the emperical correlation of the generated points on the band
points_x = A_points[0, :, 0].cpu()

e_A = e_A.squeeze()
e_corr_A = torch.corrcoef(e_A.T)

e_B = e_B.squeeze()
e_corr_B = torch.corrcoef(e_B.T)

e_C = e_C.squeeze()
e_corr_C = torch.corrcoef(e_C.T)

## finding the midpoint on the x axis of the data resolution
idx = int(25 / 2)
fig, ax = plt.subplots(1, 1)

ax.plot(points_x, E_corr_A[idx, :], label = 'Reference', color = 'black')
ax.plot(points_x, e_corr_A[idx, :], label = 'Generated (A-A)', color = 'red', linestyle = 'dashed')
ax.plot(points_x, e_corr_B[idx, :], label = 'Generated (B-B)', color = 'purple', linestyle = 'dashed')
ax.plot(points_x, e_corr_C[idx, :], label = 'Generated (C-C)', color = 'green', linestyle = 'dashed')
ax.legend()

ax.set_ylim(0.65, 1.01)
ax.set_xlabel('x1')
ax.set_ylabel('C(x1, 0.5)')
plt.show()

fig.savefig('Fig 5 (b).png')
```

```python
true_e = test_data.E.reshape(-1, resolution, resolution) 

fig, ax = plt.subplots(1, 2, figsize = (12, 5))

c = ax[0].pcolor(points_x, points_x, (true_e.mean(dim = 0) - e.mean(dim = 0)) )
fig.colorbar(c)
c = ax[1].pcolor(points_x, points_x, (true_e.std(dim = 0) - e.std(dim = 0)) )
fig.colorbar(c)

ax[0].set_title('Error of Mean Fields')
ax[0].set_xlabel('x1')
ax[0].set_ylabel('x2')

ax[1].set_title('Error of Std Fields')
ax[1].set_xlabel('x1')
ax[1].set_ylabel('x2')
fig.tight_layout()
plt.show()
```
