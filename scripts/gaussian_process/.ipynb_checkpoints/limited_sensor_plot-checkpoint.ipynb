```python
from pathlib import Path

import torch
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import pandas as pd
import numpy as np

from fgm.problems.gaussian_process import GPDataset, cov_func
from fgm.vae import FunctionalVAE
```

```python
device = torch.device(0)

use_covariance = False
bs = 1024
kld_weight = 0.005
epochs = 1000

cov_len = 0.05
variance = 0.5
```

```python
runs = {
    2: {
        'Correlation': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000),
        },
        'No Constraint': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000),
        }
    },
    3: {
        'Correlation': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000),
        },
        'No Constraint': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000),
        }
    },
    4: {
        'Correlation': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000),
        },
        'No Constraint': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000)
        }
    },
    5: {
        'Correlation': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000),
        },
        'No Constraint': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000)
        }
    },
    6: {
        'Correlation': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000),
        },
        'No Constraint': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000)
        }
    },
    7: {
        'Correlation': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000)
        },
        'No Constraint': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000),
        }
    },
    8: {
        'Correlation': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000),
        },
        'No Constraint': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000),
        }
    },
    9: {
        'Correlation': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000)
        },
        'No Constraint': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000)
        }
    },
    10: {
        'Correlation': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000)
        },
        'No Constraint': {
            0: (999, 20000),
            1: (999, 20000),
            2: (999, 20000),
        }
    }
}

df = {
    'num_sensors': [],
    'version': [],
    'Constraint': [],
    'cov_error': [],
}

for num_sensors in runs.keys():

    ## generate validation dataset
    test_data = GPDataset(10000, num_sensors, cov_len = cov_len, variance = variance)

    for constraint in runs[num_sensors].keys():
        for version in tqdm(runs[num_sensors][constraint].keys()):
            epoch, step = runs[num_sensors][constraint][version]

            if constraint == 'Correlation':
                num_colloc = 512
                res_weight = 20.0
            elif constraint == 'Correlation Weight = 100.0':
                num_colloc = 512
                res_weight = 100.0
            elif constraint == 'No Constraint':
                num_colloc = 1
                res_weight = 0.0
                
            name = f'num_sensors={num_sensors}-use_covariance={use_covariance}-bs={bs}-num_colloc={num_colloc}-res_weight={res_weight}-kld_weight={kld_weight}-epochs={epochs}'
            folder = Path(f'/Users/tashah/Desktop/functional-generative-modeling/results/gaussian_process/vae/{name}/version_{version}/')
            model_path = folder / f'checkpoints/epoch={epoch}-step={step}.ckpt'
            vae = FunctionalVAE.load_from_checkpoint(model_path).to(device)

            test_loader = DataLoader(test_data, batch_size = len(test_data), shuffle = False)
            
            y_sensor, x_sensor = next(iter(test_loader))
            y_sensor, x_sensor = y_sensor.to(device), x_sensor.to(device)
            
            x_collocation = torch.linspace(0.0, 1.0, 100).reshape(1, -1, 1).repeat(y_sensor.shape[0], 1, 1).to(device)
            
            with torch.no_grad():
                mu, logvar = vae.encode(y_sensor)
                # z = vae.reparameterize(mu, logvar)
                z = torch.randn_like(mu)
                y_collocation = vae.decode(z, x_collocation)

            cov_true = cov_func(x_collocation[0, :, :], vae.loss.cov_len, vae.loss.variance).cpu()
            cov_est = torch.cov(y_collocation.squeeze().T).cpu()

            df['num_sensors'].append(num_sensors)
            df['version'].append(version)
            df['Constraint'].append(constraint)
            df['cov_error'].append(torch.mean(torch.square(cov_true - cov_est)).item())
            
df = pd.DataFrame(df)
```

```python
fig, ax = plt.subplots(1, 1)
ax.set_xlabel('Number of Sensors')
ax.set_ylabel('Covariance Error')

sns.lineplot(df, x = 'num_sensors', y = 'cov_error', hue = 'Constraint', linestyle = 'dashed', ax = ax)
sns.scatterplot(df, x = 'num_sensors', y = 'cov_error', hue = 'Constraint', legend = False, ax = ax)
plt.show()
fig.savefig('gaussian_process.png', dpi = 600)
```

```python
num_sensors = 10
constraint = 'No Constraint'
# constraint = 'Correlation'

if constraint == 'Correlation':
    num_colloc = 512
    res_weight = 20.0
elif constraint == 'Correlation Weight = 100.0':
    num_colloc = 512
    res_weight = 100.0
elif constraint == 'No Constraint':
    num_colloc = 1
    res_weight = 0.0

version = 0

test_data = GPDataset(10000, num_sensors, cov_len = cov_len, variance = variance)
epoch, step = runs[num_sensors][constraint][version]

name = f'num_sensors={num_sensors}-use_covariance={use_covariance}-bs={bs}-num_colloc={num_colloc}-res_weight={res_weight}-kld_weight={kld_weight}-epochs={epochs}'
folder = Path(f'/Users/tashah/Desktop/functional-generative-modeling/results/gaussian_process/vae/{name}/version_{version}/')
model_path = folder / f'checkpoints/epoch={epoch}-step={step}.ckpt'
vae = FunctionalVAE.load_from_checkpoint(model_path).to(device)

test_loader = DataLoader(test_data, batch_size = len(test_data), shuffle = False)

y_sensor, x_sensor = next(iter(test_loader))
y_sensor, x_sensor = y_sensor.to(device), x_sensor.to(device)

x_collocation = torch.linspace(0.0, 1.0, 100).reshape(1, -1, 1).repeat(y_sensor.shape[0], 1, 1).to(device)

with torch.no_grad():
    mu, logvar = vae.encode(y_sensor)
    # z = vae.reparameterize(mu, logvar)
    z = torch.randn_like(mu)
    
    y_collocation = vae.decode(z, x_collocation).cpu()

cov_true = cov_func(x_collocation[0, :, :], vae.loss.cov_len, vae.loss.variance).cpu()
cov_est = torch.cov(y_collocation.squeeze().T).cpu()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 5))
ax[0].set_title('Generated Samples')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].set_ylim(-2.0, 4.0)

ax[1].set_title('Sample Covariance')
ax[1].set_xlabel('X')
ax[1].set_ylabel('X\'')

    
for i in range(500):
    ax[0].plot(x_collocation[i, :, 0].cpu(), y_collocation[i, :, 0], color = 'blue', alpha = 0.1)

sns.rugplot(x = x_sensor[0, :, 0].cpu(), ax = ax[0], label = 'Sensors', color = 'red', height = 0.2)
fig.colorbar(ax[1].pcolor(x_collocation[i, :, 0].cpu(), x_collocation[i, :, 0].cpu(), cov_est))
plt.show()

fig.savefig(f'{constraint}.png', dpi = 600)
```
