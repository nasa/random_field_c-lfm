```python
from pathlib import Path

import torch
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import pandas as pd
import numpy as np

from fgm.problems.gaussian_process import GPDataset, cov_func
from fgm.vae import FunctionalVAE
```

```python
device = torch.device(0)

num_sensors = 5
use_covariance = False
bs = 1024
num_colloc = 512
res_weight = 20.0
kld_weight = 0.005
epochs = 1000

version = 3
epoch = 999
step = 20000

name = f'num_sensors={num_sensors}-use_covariance={use_covariance}-bs={bs}-num_colloc={num_colloc}-res_weight={res_weight}-kld_weight={kld_weight}-epochs={epochs}'
folder = Path(f'/Users/tashah/Desktop/functional-generative-modeling/results/gaussian_process/vae/{name}/version_{version}/')
model_path = folder / f'checkpoints/epoch={epoch}-step={step}.ckpt'

vae = FunctionalVAE.load_from_checkpoint(model_path).to(device)
metrics = pd.read_csv(folder / 'metrics.csv')

fig, ax = plt.subplots(1, 3, figsize = (15, 5))
fig.suptitle('Log scale losses')

ax[0].set_title('Reconstruction')
ax[0].plot(np.log(metrics['reconstruction']))
ax[0].set_xlabel('Epoch')

ax[1].set_title('Residual')
ax[1].plot(np.log(metrics['residual']))
ax[1].set_xlabel('Epoch')

ax[2].set_title('KL Divergence')
ax[2].plot(np.log(metrics['kld']))
ax[2].set_xlabel('Epoch')
plt.show()
```

```python
test_data = GPDataset(10000, num_sensors, cov_len = vae.loss.cov_len, variance = vae.loss.variance)
test_loader = DataLoader(test_data, batch_size = len(test_data), shuffle = False)

y_sensor, x_sensor = next(iter(test_loader))
y_sensor, x_sensor = y_sensor.to(device), x_sensor.to(device)

x_collocation = torch.linspace(0.0, 1.0, 100).reshape(1, -1, 1).repeat(y_sensor.shape[0], 1, 1).to(device)

with torch.no_grad():
    mu, logvar = vae.encode(y_sensor)
    z = vae.reparameterize(mu, logvar)
    y_collocation = vae.decode(z, x_collocation)

y_dense, x_dense = test_data.dense_eval(range(len(test_data)))

fig, ax = plt.subplots(1, 2, figsize = (10, 3))

ax[0].set_title('Real Samples')
ax[1].set_title('Generated Samples')
ax[0].set_ylim(-3.0, 5.0)
ax[1].set_ylim(-3.0, 5.0)
for i in range(100):
    ax[0].plot(x_dense[:, 0], y_dense[i, :], color = 'blue', alpha = 0.1)
    ax[1].plot(x_collocation[i, :, 0].cpu(), y_collocation[i, :].cpu(), color = 'blue', alpha = 0.1)

sns.rugplot(x = x_sensor[0, :, 0].cpu(), ax = ax[0], label = 'Sensors', color = 'red', height = 0.2)
plt.show()
```

```python
cov_true = cov_func(x_collocation[0, :, :], vae.loss.cov_len, vae.loss.variance).cpu()
cov_est = torch.cov(y_collocation.squeeze().T).cpu()

fig, ax = plt.subplots(1, 3, figsize = (15, 4))
ax[0].set_title('True Covariance')
# c1 = ax[0].pcolor(x_collocation[0, :, 0].cpu(), x_collocation[0, :, 0].cpu(), cov_true, vmin = 0.0, vmax = 1.0)
c1 = ax[0].pcolor(x_collocation[0, :, 0].cpu(), x_collocation[0, :, 0].cpu(), cov_true)
fig.colorbar(c1, ax = ax[0])

ax[1].set_title('Generated Covariance')
c2 = ax[1].pcolor(x_collocation[0, :, 0].cpu(), x_collocation[0, :, 0].cpu(), cov_est)
fig.colorbar(c2, ax = ax[1])

ax[2].set_title('Error')
c3 = ax[2].pcolor(x_collocation[0, :, 0].cpu(), x_collocation[0, :, 0].cpu(), (cov_est - cov_true))
fig.colorbar(c3, ax = ax[2])
plt.show()
# fig.savefig('covariance.png')

torch.mean(torch.square(cov_true - cov_est))
```

```python
corr_true = cov_func(x_collocation[0, :, :], vae.loss.cov_len, vae.loss.variance).cpu() / vae.loss.variance
corr_est = torch.corrcoef(y_collocation.squeeze().T).cpu()

fig, ax = plt.subplots(1, 3, figsize = (15, 4))
ax[0].set_title('True Correlation')
c1 = ax[0].pcolor(x_collocation[0, :, 0].cpu(), x_collocation[0, :, 0].cpu(), corr_true)
fig.colorbar(c1, ax = ax[0])

ax[1].set_title('Generated Correlation')
c2 = ax[1].pcolor(x_collocation[0, :, 0].cpu(), x_collocation[0, :, 0].cpu(), corr_est)
fig.colorbar(c2, ax = ax[1])

ax[2].set_title('Error')
c3 = ax[2].pcolor(x_collocation[0, :, 0].cpu(), x_collocation[0, :, 0].cpu(), (corr_est - corr_true))
fig.colorbar(c3, ax = ax[2])
plt.show()
```
